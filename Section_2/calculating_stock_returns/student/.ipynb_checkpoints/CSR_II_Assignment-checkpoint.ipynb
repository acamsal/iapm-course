```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

ticker = "TEF.MC"
df = yf.download(ticker, start="2020-01-01", end="2025-01-01")
```

```python
df.head()
```

```python
df = df['Close']
```

```python
df.head()
```

```python
print(df.columns)
```

```python
precio_col = df.columns[0]
df = df.rename(columns={precio_col: 'price_t'})
```

```python
df.columns.name = None
```

```python
df['returns'] = (df['price_t'] / df['price_t'].shift(1)) - 1
df.head()
```

```python
df['price_t'].plot(figsize=(12,8))
```

```python
df['returns'].plot(figsize=(12,8))
```
