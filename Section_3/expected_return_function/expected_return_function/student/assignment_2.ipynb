```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

ticker = "AAPL"
df = yf.download(ticker, start="2000-01-01", end="2017-12-31", interval="1wk")
```

```python
print(df.head())
```

```python
df = df[['Close']]
df.columns = df.columns.droplevel(1)
df = df.rename(columns={'Close': 'price_t'})
df['returns'] = (df['price_t'] / df['price_t'].shift(1)) - 1
df.head()
```

```python
def getExpectedReturn(df,
                      price_col_name,
                      annualised=True,
                      annualised_method='sophisticated',
                      frequency='daily'):
    """
    Returns the expected return of a stock given price data
    """
    if frequency not in ['daily', 'weekly']:
        raise ValueError("frequency valid args are: ['daily','weekly']")

    if annualised and annualised_method not in ['sophisticated', 'crude']:
        raise ValueError("annualised_method valid args are ['sophisticated' or 'crude']")
    
    # Calculate returns of prices
    returns = df[price_col_name].pct_change(1)

    #Calculate the expected return using the mean method
    expected_return_per_period = returns.mean()

    periods_per_year = 250
    if frequency == 'weekly':
        periods_per_year = 52

    if annualised and annualised_method == 'sophisticated':
        return ((1 + expected_return_per_period) ** periods_per_year) - 1
    elif annualised and annualised_method == 'crude':
        return expected_return_per_period * periods_per_year
    else:
        return periods_per_year
```

```python
print(getExpectedReturn(df, price_col_name='price_t',
                        annualised=True,
                      annualised_method='sophisticated',
                      frequency='weekly'))
```
