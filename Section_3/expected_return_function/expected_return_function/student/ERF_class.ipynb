## Expected Returns using Average Method - Create a Function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("../data/fb_price.csv")
```

```python
def getExpectedReturn(df,
                      price_col_name,
                      annualised=True,
                      annualised_method='sophisticated'):
    """
    Returns the expected return of a stock given price data
    """
    # Calculate returns of prices
    returns = df[price_col_name].pct_change(1)

    #Calculate the expected return using the mean method
    expected_return_daily = returns.mean()

    if annualised:
        if annualised_method == 'sophisticated':
            expected_return_annual = ((1 + expected_return_daily) ** 250) - 1
        elif annualised_method == 'crude':
            expected_return_annual = expected_return_daily * 250
        return expected_return_annual
    else:
        return expected_return_daily
```

```python
print(getExpectedReturn(df, 'Adj Close'))
```

```python
print(getExpectedReturn(df, 'Adj Close', annualised=False))
```

```python
print(getExpectedReturn(df, 'Adj Close', annualised_method='crude'))
```
