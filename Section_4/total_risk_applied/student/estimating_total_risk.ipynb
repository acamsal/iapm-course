## Estimating the Total Risk of a Stock II

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/fb_price.csv")
df = df[['Date', 'Adj Close']]
df = df.rename(columns={'Date': 'date',
                        'Adj Close': 'price_t'})
df['returns'] = df['price_t'].pct_change(1)
df.head()
```

```python
#ddof to do 1/(n+1) instead of 1/(n+0)
var_fb = np.var(df['returns'], ddof=1)
print(var_fb)
```

```python
sd_fb = np.sqrt(var_fb)
print(sd_fb)
```

```python
print(np.std(df['returns'], ddof=1))
```

```python
df.head()
```

```python
df['deviations'] = df['returns'] - df['returns'].mean()
df.head()
```

```python
df['squared_deviations'] = df['deviations'] ** 2
df.head()
```

```python
#sum(df['squared_deviations'].dropna())
print(np.sum(df['squared_deviations'])) #drops the nans by default
```

```python
ssd = np.sum(df['squared_deviations'])
var_fb_manual =  ssd / (len(df['squared_deviations'].dropna()) - 1)
print(var_fb_manual, var_fb_manual == var_fb)
```

```python
sd_fb_manual = np.sqrt(var_fb_manual)
print(sd_fb_manual, sd_fb_manual == sd_fb)
```

```python
sd_fb_annual = sd_fb * np.sqrt(250)
print(sd_fb_annual)
```

```python
returns_annual_crude = df['returns'].mean() * 250
print(returns_annual_crude)
```

```python
returns_annual_soph = ((1 + df['returns'].mean()) ** 250) - 1
print(returns_annual_soph)
```
