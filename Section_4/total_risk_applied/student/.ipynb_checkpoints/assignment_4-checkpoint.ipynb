```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

ticker = "AAPL"
df = yf.download(ticker, start="2000-01-01", end="2005-01-01")
```

```python
df = df[['Close']]
df.columns = df.columns.droplevel(1)
df = df.rename(columns={'Close': 'price_t'})
df['returns'] = (df['price_t'] / df['price_t'].shift(1)) - 1
df.head()
```

```python
var_aapl = np.var(df['returns'], ddof=1)
sd_aapl = np.sqrt(var_aapl)
print("Daily Variance of AAPL:", var_aapl)
print("Daily Standard Deviation of AAPL:", sd_aapl)
```

```python
annual_sd_aapl = sd_aapl * np.sqrt(250)
print("Annual Standard Deviation of AAPL:", annual_sd_aapl)
```
