We have to set the date as an index because, from now on, we are going to ask Python to run calculations on the entire dataframe

```python
returns = df.pct_change(1)
returns.dropna(inplace=True)
```

```python
returns.head()
```

```python
num_stocks = 10
weights = [1 / num_stocks] * num_stocks
```

```python
weights
```

```python
vcv_matrix = returns.cov(ddof=1)
```

```python
vcv_matrix
```

```python
var_p = np.dot(np.transpose(weights),
               np.dot(vcv_matrix, weights))
```

```python
var_p
```

```python
sd_p = np.sqrt(var_p)
```

```python
sd_p
```

```python
sd_p_annual = sd_p * np.sqrt(250)
```

```python
sd_p_annual
```

```python
individual_risks = returns.std(ddof=1) * np.sqrt(250)
```

```python
print(individual_risks)
```

Just by having an equally weighted portfolio, we outperformed the risk of every single stock
