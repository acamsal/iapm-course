```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

tickers = ['NVDA', 'AAPL', 'GOOG', 'MSFT', 'AMZN',
           'META', 'AVGO', 'TSLA', 'WMT', 'LLY']
prices = yf.download(tickers, start="2020-01-01", end="2025-01-01")['Close']
```

```python
prices.dropna(inplace=True)
```

```python
returns = prices.pct_change(1)
returns.dropna(inplace=True)
returns.head()
```

```python
num_stocks = 10
weights = [1 / num_stocks] * num_stocks
print(weights)
```

```python
vcv_matrix = returns.cov(ddof=1)
vcv_matrix
```

```python
var_p = np.dot(np.transpose(weights),
               np.dot(vcv_matrix, weights))
sd_p = np.sqrt(var_p)
print("Portfolio variance:", var_p)
print("Portfolio standard variation:", sd_p)
```
