## Optimizing Portfolio Weights - Achieving a Target Return

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
```

```python
df = pd.read_csv('../data/15stocks_price.csv')
df.set_index('date_gsheets', inplace=True)
```

```python
returns = df.pct_change(1).dropna()
```

minimize(fun, x0, bounds=None, constraints=())

fun: the objective function to be minimized

x0: initial guess (weights of the initial securities)

bounds: weights of individual securities must be between 0 and 1

constraints: sum of individual weights is always equal to 1, we also specify the difference between the expected return of the portfolio and the target return

```python
def getPortReturn(weights):
    exp_ret_portfolio = np.dot(np.transpose(weights),
                              returns.mean()) * 250
    return exp_ret_portfolio
```

```python
num_stocks = len(returns.columns)
init_weights = [1 / num_stocks] * num_stocks
```

```python
getPortReturn(init_weights)
```

```python
target_return = 0.4
```

```python
bounds = tuple((0,1) for i in range(num_stocks))
```

```python
#the constraint is if the value of the function within the
# constraint returns zero
constraints = ({'type' : 'eq',
                'fun' : lambda w : np.sum(w) - 1},
              {'type' : 'eq',
               'fun' : lambda w : w.dot(returns.mean()) * 250 - target_return})
```

```python
results = minimize(fun=getPortReturn,
                  x0=init_weights,
                  bounds=bounds,
                  constraints=constraints)
```

```python
results
```

```python
print(getPortReturn(weights=results['x']))
```

```python
optimized_weights = pd.DataFrame(results['x'], columns=['weight'])
```

```python
optimized_weights.index = returns.columns
```
